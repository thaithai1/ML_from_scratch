```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

## I. Dataset

```python
np.random.seed(5)

sample_size=100
X=np.random.rand(sample_size)*10
Y=np.sqrt(X)
```

```python
plt.title("Y against X")
plt.scatter(X,Y)
plt.show()
```

```python
idx=np.random.permutation(sample_size)
X_train,X_test = X[idx[:int(sample_size*0.8)]],X[idx[int(sample_size*0.8):]]
Y_train,Y_test = Y[idx[:int(sample_size*0.8)]],Y[idx[int(sample_size*0.8):]]

plt.figure(figsize=(20,10))
plt.plot(X_train,Y_train,'c+',label='Training')
plt.plot(X_test,Y_test,'r+',label='Test')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
