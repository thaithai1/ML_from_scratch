```python
from Model import LinearReg
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
%load_ext autoreload
%autoreload 2
```

## I. Dataset

```python
sample_size=1000
np.random.seed(1)
X=(np.random.rand(sample_size)*10).reshape(-1,1).T
Y=np.sqrt(X).reshape(-1,1).T

plt.figure(figsize=(15,5))
plt.title("Y against X")
plt.scatter(X,Y)
plt.show()
```

```python
#Train set and Test set
idx=np.random.permutation(sample_size)
X_train,X_test = X[:,idx[:int(sample_size*0.8)]],X[:,idx[int(sample_size*0.8):]]
Y_train,Y_test = Y[:,idx[:int(sample_size*0.8)]],Y[:,idx[int(sample_size*0.8):]]

plt.figure(figsize=(15,5))
plt.plot(X_train[0,:],Y_train[0,:],'c+',label='Training')
plt.plot(X_test[0,:],Y_test[0,:],'r+',label='Test')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

## II.Linear Regression

```python
model=LinearReg(X_train, Y_train, X_test, Y_test)
model.learning_curve_ex(iter = 1000, lr = 0.01, m_step = 1, max_ex = 10)
```

```python
np.random.seed(1)
model=LinearReg(X_train, Y_train, X_test, Y_test)
model.train( iter=100, lr=0.0001, verbose=True)
```

**Training using annealing learning rate**

```python
np.random.seed(1)
model=LinearReg(X_train, Y_train, X_test, Y_test)
model.train_ann_lr( iter = 100, verbose=True)
```

## III. Model Vizualisation

```python
X=np.arange(0,np.max(X_train),0.2).reshape(1,-1)
Y=model.predict(X)
plt.figure(figsize=(15,8))
plt.plot(X.reshape(-1,1),Y.reshape(-1,1), label = 'Model')
plt.plot(X_test.reshape(-1,1),Y_test.reshape(-1,1),'r+', label = 'Test set')
plt.legend(bbox_to_anchor=(1.0, 1), loc=2, borderaxespad=0.)
plt.show()
```
